```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.isna().sum()
```

## Taking care of missing data

- delete or replace  Nan or blank values 

```python
age_mean = df['Age'].mean()
```

```python
df['Age'].fillna(value=age_mean, inplace=True) 
```

```python
df.isna().sum()
```

## Encoding the Dependent Variable

- Convert categorical data into numeric data for further calculation for our Y axies or our predicted values 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
df['Sex_EN'] = le.fit_transform(df['Sex'])
```

## Encoding the Independent Variable

- Convert categorical data into numeric data for further calculation for our X axies or our Features values  

```python
df = pd.get_dummies(df, columns=['Embarked'], prefix=['Embarked']) # Chnage any Prefix
```

```python
df
```

## Feature Scaling | Normalization

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler # Convert into lowe number like 0 to 1 
#  Standardization
from sklearn.preprocessing import StandardScaler # convert into lower number like -1 to 1
scaler = MinMaxScaler()
scaled_data_age = scaler.fit_transform(df[['Age']])
scaled_data_fare = scaler.fit_transform(df[['Fare']])
```

```python
df[['Age']]= scaled_data_age
df[['Fare']]= scaled_data_fare
```
